```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = 'datasets/rgbd_dataset_freiburg2_desk/groundtruth.txt'
data = np.loadtxt(file_path, comments='#')

#So the GT is the absolute pose (absolute position..)


# Extract the timestamp, translation (tx, ty, tz), and orientation (qx, qy, qz, qw)
timestamps = data[:, 0]
translations = data[:, 1:4]
orientations = data[:, 4:8]

# Extract tx and ty
tx = translations[:, 0]
ty = translations[:, 1]

# Plot the trajectory
plt.figure()
plt.plot(tx, ty, label='Trajectory')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('SLAM Trajectory')
plt.legend()
plt.show()
```
